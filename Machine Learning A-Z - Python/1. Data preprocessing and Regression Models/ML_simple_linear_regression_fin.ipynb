# Simple Linear Regression

## Importing the libraries

```
import numpy as np #to process arrays as inputs. "as np" to make a shortcut call in case we want to use numpy
import matplotlib.pyplot as plt #to plot charts. "." to use specific module pyplot and "plt" as shortcut
import pandas as pd #import data set and matrices. "pd" -> shortcut
```

## Importing the dataset

```
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```
from sklearn.linear_model import LinearRegression #importar classe LinearRegression
regressor = LinearRegression() #criar instancia
regressor.fit(X_train, y_train) #fit method --> treina o regression model
```

## Predicting the Test set results

```
y_pred = regressor.predict(X_test) #predict dos dados testes com base no modelo anterior.
#Posteriormente iremos auferir a diferença do predict com o real
#y_pred fica com o predicted to modelo. Agora vamos comparar com os numeros reais dos testes
```

## Visualising the Training set results

```
plt.scatter(X_train, y_train, color = 'red') #plt é o plotmatlib. Queremos ver os reais dos dados de treino em dots vermelhos
plt.plot(X_train, regressor.predict(X_train), color = 'blue') #vamos fazer a regressao para os dados treino e ver também o que deu-
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```
plt.scatter(X_test, y_test, color = 'red') #plt é o plotmatlib. Queremos ver os reais dos dados de treino em dots vermelhos
plt.plot(X_train, regressor.predict(X_train), color = 'blue') #podiamos alterar para x_test mas a linha de regressao é igual
#portanto alterar ou nao é igual.
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
